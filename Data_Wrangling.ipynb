```python
import pandas as pd
```

```python
Trips_2013 = pd.read_csv('Divvy_Trips_2013.csv')
Trips_2013['Trip_Start'] = pd.to_datetime(Trips_2013['starttime'])
Trips_2013['date'] = Trips_2013['Trip_Start'].dt.date
Trips_2013['id'] = Trips_2013['from_station_id']
```

```python
DAT_Station = Trips_2013.groupby(['date','id']).count()['tripduration']
DAT_Station = pd.DataFrame(DAT_Station)
DAT_Station.reset_index(level=0, inplace=True)

#DAT = Trips_2013.groupby(['date']).count()['tripduration']
#DAT = pd.DataFrame(DAT)
#DAT.reset_index(level=0, inplace=True)
```

```python
Stations_2013 = pd.read_csv('Divvy_Stations_2013.csv')[['id','name','latitude','longitude']]
Final_Data = pd.merge(DAT_Station, Stations_2013, how='inner', on='id')
Final_Data = Final_Data.sort_values(by=['date','id'])
```

```python
Final_Data.to_csv("Divvy_Trips_2013_Processed.csv", encoding='utf-8', index=False)
#DAT.to_csv("DV_Data_Temp.csv", encoding='utf-8', index=False)
```
